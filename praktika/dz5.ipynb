```python

tst.dropna(axis=0, subset=['Embarked'], inplace=True)
tst['Age'] = tst.groupby(['Pclass','Sex'])['Age'].transform(lambda x: x.fillna(x.median()))
tst['Fare'] = tst.groupby(['Pclass','Sex'])['Fare'].transform(lambda x: x.fillna(x.median()))
le = preprocessing.LabelEncoder()
tst['Embarked'].replace({'S':0, 'C':1, 'Q':2}, inplace=True)
tst['Sex'].replace({'male':1, 'female':0}, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

y = trn['Survived']
X = trn.drop(['Survived'], axis=1)


# Divide data into training and validation subsets
X_train, X_val, y_train, y_val = train_test_split(X, y,  test_size=0.3,
                                                           random_state=42)
```

```python
import seaborn as sns
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.svm import SVC
m1 = LogisticRegression(solver='liblinear', random_state=42)
m2 = GradientBoostingClassifier()
m3 = RandomForestClassifier()
m4 = SGDClassifier()
m5 = SVC()

models = [ m1, m2, m3, m4, m5]
i = 0
for model in models:
    i +=1
    model.fit(X_train, y_train)
    prediction = model.predict(X_val)
    print(accuracy_score(y_val, prediction))
```

```python
model = GradientBoostingClassifier(min_samples_split=20, min_samples_leaf=60, max_depth=3, max_features=7)
model.fit(X_train, y_train)
prediction = model.predict(X_val)
print(accuracy_score(y_val, prediction))
```
